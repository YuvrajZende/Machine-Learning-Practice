##ID3 and GINI

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/pacific.csv")
data.head(10)
```

```python
data.info()
```

```python
def entrpoy(y):
  values, counts = np.unique(y, return_counts=True)
  probs = counts/counts.sum()
  return -np.sum(probs*np.log2(probs))

def information_gain(data,feature,target):
  total_entropy = entrpoy(data[target])
  values, counts = np.unique(data[feature], return_counts=True)
  weight =  0
  for v,c in zip(values,counts):
    subset = data[data[feature] == v]
    weight += (c/counts.sum()) * entrpoy(subset[target])
  return total_entropy - weight
```

###GINI Index  (CART)

```python
def gini(y):
  values, counts = np.unique(y, return_counts=True)
  probs = counts/counts.sum()
  return 1 - (probs**2).sum()

def gini_index(data,feature,target):
  values,count = np.unique(data[feature], return_counts=True)
  gini_tot = 0
  for v,c in zip(values,count):
    subset = data[data[feature] == v]
    gini_tot += (c/count.sum()) * gini(subset[target])
  return gini_tot
```

Feature Selection

```python
target = data.columns[-1]
features = data.columns[:-1]
print("\n Entropy \n")
for f in features:
  print(f,entrpoy(target))

print("\n Information Gain \n")
for f in features:
  print(f,information_gain(data,f,target))

print("\n Gini Index (CART) \n")
for f in features:
  print(f,gini_index(data,f,target))
```
