```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv('/content/titanic - titanic.csv')

feactures = ['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']

df = df[feactures + ['Survived']].dropna()

print(df.head(10))
print("\n")
print("\n")
print("\n")
df.info()
```

```python
num_feactures = ['Age','SibSp','Parch','Fare']
cat_feactures = ['Pclass','Sex','Embarked']

#transformer
num_transformer = StandardScaler()
cat_transformer = OneHotEncoder(handle_unknown='ignore')

#combine transformer into a preprocessor things
preprocessor = ColumnTransformer([
     ('num',num_transformer,num_feactures),
     ('cat',cat_transformer,cat_feactures)
 ])
```

```python
# now we will split our data into X and Y
X = df[feactures]
Y = df['Survived']

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)

pipeline = Pipeline([
    ('preprocessor',preprocessor),
    ('classifier',RandomForestClassifier(n_estimators=100, random_state=42))
])

pipeline = pipeline.fit(X_train,Y_train)
print("Model training Completed")
print("\n")
y_prediction = pipeline.predict(X_test)
accuracy = accuracy_score(Y_test,y_prediction)
print("Accuracy = ",accuracy)
```

```python
sample_data = pd.DataFrame([{
    'Pclass': 3,
    'Sex':'male',
    'Age': 25,
    'SibSp':1,
    'Parch':0,
    'Fare': 7.25,
    'Embarked':'S'
}])

prediction = pipeline.predict(sample_data)
print(prediction)
if(prediction[0]==1):
  print("Survived")
else:
  print("Not Survived")
```
